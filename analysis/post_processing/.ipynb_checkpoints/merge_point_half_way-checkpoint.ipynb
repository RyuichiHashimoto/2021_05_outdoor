```python
import pandas as pd

from lib.io import load_pickle_data
from lib.noglobal import noglobal

from external_lib.evaluation_function import calc_haversine,evaluate_function

import c_library 
```

```python










#baseline
train_df_path = "/work/data/input/selfmade_dataset/baseline_with_derived_data_v5/train.pkl"
gt = load_pickle_data(train_df_path)

## best score

train_df_path = "/work/data/submission_list/baseline_with_snap_to_grid_to_2_or_3_kalman_interplorate_patial_twice/train_result.csv"
train_df = pd.read_csv(train_df_path)

train_df["latDeg_gt"] = gt["latDeg_gt"]
train_df["lngDeg_gt"] = gt["lngDeg_gt"]
```

```python


## True: to be merged
## False: not to be merged
@noglobal()
def merge_judge_SJC(df,n_of_data=100,threshold=8):
    
    pd_list = [];
    
    for key,each_df in df.groupby(["collectionName","phoneName"]):
        tmp_df = each_df.copy();
                
        
        num = tmp_df[["latDeg","lngDeg"]].to_numpy().tolist();
        s =  c_library.merge_points_on_the_road(num,n_of_data,threshold)
        tmp_df[["latDeg","lngDeg"]] = s
            
        pd_list.append(tmp_df)
        
    ret_df = pd.concat(pd_list).sort_index();
        
    
    return ret_df
```

```python

tmp_df = train_df.copy();
train_df["dist"] = calc_haversine(train_df["latDeg"],train_df["lngDeg"],train_df["latDeg_gt"],train_df["lngDeg_gt"])
print("before",evaluate_function(train_df,"dist"))

pd_list = [];
for key,each_df in tmp_df.groupby(["collectionName","phoneName"]):
    tmp_each_df = each_df.copy();
    
    ret_df_per_path = merge_judge_SJC(tmp_each_df,3,1.1);
    #ret_df_per_path = merge_judge_SJC(ret_df_per_path,5,0.5);
    #ret_df_per_path = merge_judge_SJC(ret_df_per_path,10,0.25);
    
    ret_df_per_path["dist"] = calc_haversine(ret_df_per_path["latDeg"],ret_df_per_path["lngDeg"],ret_df_per_path["latDeg_gt"],ret_df_per_path["lngDeg_gt"])
    each_df["dist"] = calc_haversine(each_df["latDeg"],each_df["lngDeg"],ret_df_per_path["latDeg_gt"],ret_df_per_path["lngDeg_gt"])
    #print(key[0]+"-"+key[1]+"\t\t",each_df["dist"].mean(),ret_df_per_path["dist"].mean())
    
    pd_list.append(ret_df_per_path);
    
ret_df = pd.concat(pd_list,axis=0).sort_index()
ret_df["dist"] = calc_haversine(ret_df["latDeg"],ret_df["lngDeg"],train_df["latDeg_gt"],train_df["lngDeg_gt"])
print("after",evaluate_function(ret_df,"dist"))

    
```

```python
for key,each_df in ret_df.groupby(["collectionName","phoneName"]):
    

    tmp_train_df = train_df[(train_df["phone"] == key[0]+"_"+key[1] )]        
    tmp_train_df["dist"] = calc_haversine(tmp_train_df["latDeg"],tmp_train_df["lngDeg"],train_df["latDeg_gt"],train_df["lngDeg_gt"])
    if (tmp_train_df["dist"].mean() > each_df["dist"].mean()):
        print("-------------",key[0],key[1],"-------------")        
        print("brefore",tmp_train_df["dist"].mean())
        print("after",each_df["dist"].mean())
        

    
```
