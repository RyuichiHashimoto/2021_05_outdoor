```python
import pandas as pd
import json
```

```python
path  =   "/work/data/input/google-smartphone-decimeter-challenge/google-sdc-corrections-main/osr/json/2020-05-14-US-MTV-2.json"
#df = json.load(open(path,"r"))

res = []
decoder = json.JSONDecoder()
with open(path, 'r') as f:
    line = f.readline()
    while line:
        res.append(decoder.raw_decode(line))
        line = f.readline()
```
