```python

# from https://www.kaggle.com/sohier/loading-gnss-logs
def gnss_log_to_dataframes(path):
    
    gnss_section_names = {'Raw','UncalAccel', 'UncalGyro', 'UncalMag', 'Fix', 'Status', 'OrientationDeg'}
    with open(path) as f_open:
        datalines = f_open.readlines()

    datas = {k: [] for k in gnss_section_names}
    gnss_map = {k: [] for k in gnss_section_names}
    for dataline in datalines:
        is_header = dataline.startswith('#')
        dataline = dataline.strip('#').strip().split(',')
        # skip over notes, version numbers, etc
        if is_header and dataline[0] in gnss_section_names:
            gnss_map[dataline[0]] = dataline[1:]
        elif not is_header:
            datas[dataline[0]].append(dataline[1:])

    results = dict()
    for k, v in datas.items():
        results[k] = pd.DataFrame(v, columns=gnss_map[k])
    # pandas doesn't properly infer types from these lists by default
    for k, df in results.items():
        for col in df.columns:
            if col == 'CodeType':
                continue
            results[k][col] = pd.to_numeric(results[k][col])

    return results
```

```python

def visualize_trafic(df,color_header:str=None, zoom=9,outputfile=None):
    

    center = {"lat":(df["latDeg"].max()  + df["latDeg"].min())/2, "lon": (df["lngDeg"].max()  + df["lngDeg"].min())/2}
    
    
    if (color_header == None):
        sample_ground_truth_df["history"] =  [ i for i in range(sample_ground_truth_df.shape[0])]
        color_header = "history"
    
    fig = px.scatter_mapbox(df,
                            
                            # Here, plotly gets, (x,y) coordinates
                            lat="latDeg",
                            lon="lngDeg",
                            
                            #Here, plotly detects color of series
                            color=color_header,
                            labels="phoneName",
                            
                            zoom=zoom,
                            center=center,
                            height=600,
                            width=800)
    fig.update_layout(mapbox_style='stamen-terrain')
    fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
    fig.update_layout(title_text="GPS trafic")
    
    
    if (not outputfile == None):
        if (outputfile.endswith(".html")):
            fig.write_html(outputfile);
        elif(outputfile.endswith(".png")):
            fig.write_image(outputfile,format = "png")
        else:
            raise Exception("Sorry, I cannot understand the " + outputfile.split(".")[-1] +" file");
    else:
        return fig
    
```

```python
def add_unixtime_from_gpstime(df):    
    copied_df = df.copy();
    copied_df["unix_timestamp"] = copied_df["millisSinceGpsEpoch"].map(lambda x: (datetime(1980, 1, 6) + timedelta(seconds=round(x)/1000 - (35 - 19))).timestamp());
    copied_df["unix_timestamp_date"] = copied_df["unix_timestamp"].map(lambda x: datetime.fromtimestamp(x));
    return copied_df
    
    
```

### UncalGyro

```python
Uncal_gyro = sample_gnss_df_set["UncalGyro"];
```

```python
Uncal_gyro.columns.values
```

#### gyro x, y, zの推移データのグラフプロット

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(30, 12))

y_limit_threshhold = 0.6

Uncal_gyro.set_index("utcTimeMillis")["UncalGyroXRadPerSec"].plot(ax=axes[0][0],title="Ungyro_x",ylim=[-1*y_limit_threshhold,y_limit_threshhold]);
Uncal_gyro.set_index("utcTimeMillis")["UncalGyroYRadPerSec"].plot(ax=axes[0][1],title="Ungyro_y",ylim=[-1*y_limit_threshhold,y_limit_threshhold]);
Uncal_gyro.set_index("utcTimeMillis")["UncalGyroZRadPerSec"].plot(ax=axes[0][2],title="Ungyro_z",ylim=[-1*y_limit_threshhold,y_limit_threshhold]);

Uncal_gyro.set_index("utcTimeMillis")["UncalGyroXRadPerSec"].rolling(10).mean().plot(ax=axes[1][0],title="rolling Ungyro_x",ylim=[-1*y_limit_threshhold,y_limit_threshhold]);
Uncal_gyro.set_index("utcTimeMillis")["UncalGyroYRadPerSec"].rolling(10).mean().plot(ax=axes[1][1],title="rolling Ungyro_y",ylim=[-1*y_limit_threshhold,y_limit_threshhold]);
Uncal_gyro.set_index("utcTimeMillis")["UncalGyroZRadPerSec"].rolling(10).mean().plot(ax=axes[1][2],title="rolling Ungyro_z",ylim=[-1*y_limit_threshhold,y_limit_threshhold]);
```

#### データ取得間隔

以下を見る限り、最大で6ms間隔でデータを取得していることがわかる。

```python
Uncal_gyro["utcTimeMillis"].diff().dropna().unique()
```
